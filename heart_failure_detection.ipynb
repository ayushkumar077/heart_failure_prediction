```python
X=df.iloc[:,0:-1]
y=df['HeartDisease']
```

```python
chestpaintype=pd.get_dummies(X['ChestPainType'],drop_first=True,dtype=int)
sex=pd.get_dummies(X['Sex'],drop_first=True,dtype=int)
exercise=pd.get_dummies(X['ExerciseAngina'],drop_first=True,dtype=int)
```

```python
X.drop(columns=['Sex','ChestPainType','ExerciseAngina'],inplace=True)
```

```python
X=pd.concat([X,sex,chestpaintype,exercise],axis=1)
```

```python
X.head()
```

```python
X=X.drop(columns=['ST_Slope','RestingECG'])
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import ReLU,Dropout
```

```python
model=Sequential()
```

```python
model.add(Dense(units=11,activation='relu'))
model.add(Dense(units=6,activation='relu'))
model.add(Dense(units=7,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,validation_split=0.30,batch_size=100,epochs=300)
```

```python
import numpy as np
y_pred=model.predict(X_test)
y_pred_class=(y_pred>=0.50).astype(int)
```

```python
y_pred_class
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred_class,y_test)
```
